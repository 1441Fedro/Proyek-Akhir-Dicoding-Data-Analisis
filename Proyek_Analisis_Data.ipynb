## Visualization & Explanatory Analysis

### Pertanyaan 1: Berapa rata-rata tahunan jumlah NO2 di setiap kota?

```python
question1_df = all_df.groupby(by=["station"]).agg({
    "NO2_mean": "mean"
}).sort_values(by="NO2_mean", ascending=False)

question1_df.rename(columns={
    "NO2_mean": "Rata-rata Tahunan"
}, inplace=True)

plt.figure(figsize=(10, 8))

sns.barplot(
    y="Rata-rata Tahunan",
    x="station",
    data=question1_df.sort_values(by="Rata-rata Tahunan", ascending=False),
)
for i, value in enumerate(question1_df["Rata-rata Tahunan"]):
    plt.text(i, value, str(round(value, 3)), ha='center', va='bottom')
plt.title("Rata - Rata Tahunan NO2 di Beberapa Kota", loc="center", fontsize=15)
plt.ylabel(None)
plt.yticks([0,5,10,15,20,25,30,35,40,45,50,55,60,65,70])
plt.xlabel("Nama Kota")
plt.tick_params(axis='x', labelsize=10, rotation=46)
plt.show()
```

### Pertanyaan 2: Pada tahun berapa dan di kota mana jumlah seluruh NO2 dalam 1 tahun paling tinggi?

```python
question2_df = all_df.groupby(by=["station", "year"]).agg({
    "NO2_sum": "sum"
})

question2_df.rename(columns={
    "NO2_sum": "Jumlah Tahunan"
}, inplace=True)

plt.figure(figsize=(20, 8))

sns.barplot(data=question2_df, x="year", y="Jumlah Tahunan", hue="station", errorbar=None)
plt.show()
```
